# load data 

```python
import numpy as np

docs = []
for i in range(1, 60):
    name = 'dataset\\' + str(i)+'.txt'
    with open(name) as f:
        x = f.read()
        
        x = x.replace('\n', ' ')
        docs.append(x)
        f.close()
docs[0]

```

# inerted index


```python
unique_terms = {term for doc in docs for term in doc.split()}
unique_terms
```

```python
# Construct an inverted index

inverted_index = {}

for i, doc in enumerate(docs):

    for term in doc.split():
        if term in inverted_index:
            inverted_index[term].add(i)
        else:
            inverted_index[term] = {i}

inverted_index
```

```python
# test with computer, netflix , car , Anime, manga, Glassdoor, programming
posting_list = inverted_index['programming']
posting_list
```

```python
def or_postings(posting1, posting2):
    # the prameters are lists
    p1 = 0
    p2 = 0
    result = list()
    while p1 < len(posting1) and p2 < len(posting2):

        if posting1[p1] == posting2[p2]:
            result.append(posting1[p1])
            p1 += 1
            p2 += 1
        elif posting1[p1] > posting2[p2]:
            result.append(posting2[p2])
            p2 += 1
        else:
            result.append(posting1[p1])
            p1 += 1

    while p1 < len(posting1):
        result.append(posting1[p1])
        p1 += 1
    while p2 < len(posting2):
        result.append(posting2[p2])
        p2 += 1
    return result
```

```python
pl_1 = list(inverted_index['Anime'])
pl_2 = list(inverted_index['netflix'])
or_postings(pl_1, pl_2)
```

```python
def and_postings(posting1, posting2):
    p1 = 0
    p2 = 0
    result = list()
    while p1 < len(posting1) and p2 < len(posting2):
        if posting1[p1] == posting2[p2]:
            result.append(posting1[p1])
            p1 += 1
            p2 += 1
        elif posting1[p1] > posting2[p2]:
            p2 += 1
        else:
            p1 += 1
    return result
```

```python
pl_1 = list(inverted_index['company'])
pl_2 = list(inverted_index['Glassdoor'])

and_postings(pl_1, pl_2)
```

# TF-IDF


```python
# bag of words for each doc/file

bagofwords_list = []
cnt = 0
for doc in docs:
    bagofwords = doc.split(' ')
    if len(doc)<100:
        print(bagofwords)
    bagofwords_list.append(bagofwords)
    cnt+=1

```

```python
# count each word in all docs

numofword_list = []  # list of dictionaries
for i in range(11):
    temp = dict.fromkeys(unique_terms, 0)
    numofword_list.append(temp)

numofword_list[10]
```

```python
for i in range(11):
    bagofwords = bagofwords_list[i]
    for word in bagofwords:
        if word in numofword_list[i].keys():
            numofword_list[i][word] += 1
numofword_list[10]
```

```python
def computeTF(wordDict, bagOfWords):
    tfDict = {}
    bagOfWordsCount = len(bagOfWords)
    for word, count in wordDict.items():
        tfDict[word] = count / float(bagOfWordsCount)
    return tfDict
```

```python
TF_list = []
for i in range(10):
    tf = computeTF(numofword_list[i], bagofwords_list[i])
    TF_list.append(tf)
    
TF_list[0]
```

```python
def computeIDF(documents):
    import math
    N = len(documents)

    idfDict = dict.fromkeys(documents[0].keys(), 0)
    for document in documents:
        for word, val in document.items():
            if val > 0:
                idfDict[word] += 1

    for word, val in idfDict.items():
        if val ==0:
            val = .1
            
        idfDict[word] = math.log(N / float(val))
    return idfDict
```

```python
IDFs = computeIDF(numofword_list)
IDFs
```

```python
def computeTFIDF(tfBagOfWords, idfs):
    tfidf = {}
    for word, val in tfBagOfWords.items():
        tfidf[word] = val * idfs[word]
    return tfidf
```

```python
TF_IDF_list = []
for i in range(10):
    temp = computeTFIDF(TF_list[i], IDFs)
    TF_IDF_list.append(temp)
TF_IDF_list
```

```python
import pandas as pd
df = pd.DataFrame(TF_IDF_list)
df.head()
```
